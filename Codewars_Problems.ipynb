### **Problem Statement**
#### If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23. Finish the solution so that it returns the sum of all the multiples of 3 or 5 below the number passed in. Additionally, if the number is negative, return 0. Note: If a number is a multiple of both 3 and 5, only count it once.

```julia
function sumofmultiples(number)
  if number <= 0
    return 0
  else
    list = collect(1:number-1)
    multiples = Int64[]
    for i in list
      if i%3 == 0
        push!(multiples,i)
      elseif i%5 == 0
        push!(multiples,i)
      end
    end
  end
  return(sum(multiples))
end
```

```julia
## Testing the function

num = 10

return_sum = sumofmultiples(num)
```
